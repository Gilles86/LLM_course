```python
import numpy as np
import torch
```

# Two calculations on two separate sets of vectors

```python
# create two activation vectors
acts_1 = np.arange(0,5)
acts_2 = np.arange(2,7)

# and two weights vectors
weights_1 = np.linspace(-1,.4,len(acts_1))
weights_2 = np.linspace(-.4,2,len(acts_2))**4

# their dot products
dp_1 = np.sum(acts_1*weights_1)
dp_2 = np.sum(acts_2*weights_2)

print(dp_1,dp_2)
```

```python
# and the cross-terms!
dp_3 = np.sum(acts_1*weights_2)
dp_4 = np.sum(acts_2*weights_1)

print(dp_3,dp_4)
```

# Combine into matrices

```python
acts = np.array([acts_1,acts_2])
weights = np.array([weights_1,weights_2])

print(acts,'\n')
print(weights)
```

```python
# check the sizes
acts.shape, weights.shape
```

```python
# using matrix multiplication
acts @ weights.T
```

```python
np.dot(acts,weights.T)
```

# Dot products from Hadamard multiplication

```python
# Hadamard, aka element-wise, multiplication
acts * weights
```

```python
np.sum( acts*weights, axis=1)
```

```python
# but no cross-terms :(
```

# In PyTorch

```python
actsTorch = torch.tensor(acts).to(torch.float64)
weightsTorch = torch.tensor(weights)

actsTorch @ weightsTorch.T
```

```python
torch.matmul(actsTorch,weightsTorch.T)
```
