```python
import numpy as np
import matplotlib.pyplot as plt

# statsmodels to run the regression
import statsmodels.api as sm
import scipy.stats as stats

# vector plots
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

# Comparison in simulations

```python
# generate data
samplesize = 30

verbMeans = np.linspace(-2,3,51)

lr = np.zeros((len(verbMeans),2))
tt = np.zeros((len(verbMeans),2))

for idx,vmean in enumerate(verbMeans):

  ## create the data
  actNouns = np.random.normal(loc=0,size=samplesize)
  actVerbs = np.random.normal(loc=vmean,size=samplesize)


  ## logistic regression
  logreg = sm.Logit(
      np.hstack((np.zeros(samplesize),np.ones(samplesize))),
      sm.add_constant(np.hstack((actVerbs,actNouns)))
      ).fit(disp=False)

  lr[idx,0] = logreg.params[1]
  lr[idx,1] = -np.log(logreg.pvalues[1])


  ## t-test
  ttest = stats.ttest_rel(actNouns,actVerbs)
  tt[idx,0] = ttest.statistic
  tt[idx,1] = -np.log(ttest.pvalue)
```

```python
_,axs = plt.subplots(1,2,figsize=(10,5))

axs[0].plot(lr[:,0],tt[:,0],'ko',markerfacecolor=[.7,.9,.7,.6],markersize=9)
axs[0].set(xlabel='Logistic regression $\\beta$',ylabel='T-value',title='Parameters')

axs[1].plot(lr[:,1],tt[:,1],'ks',markerfacecolor=[.7,.7,.9,.6],markersize=9)
axs[1].axhline(-np.log(.05),linestyle='--',linewidth=.4,color='k')
axs[1].axvline(-np.log(.05),linestyle='--',linewidth=.4,color='k')
axs[1].set(xlabel='Logistic regression -ln(p)',ylabel='T-test -ln(p)',title='Statistical significance (p-values)')

plt.suptitle('Logistic regression vs. t-test',fontweight='bold',fontsize=18)
plt.tight_layout()
plt.show()
```
